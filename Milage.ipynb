<b>Step 1: Convert Trans, Drive, Fuel, Veh Class, SmartWay column values to integers.</b><br/>

First we will import pandas and read the csv file.

```python
import pandas as pd
df = pd.read_csv("milage.csv", skipinitialspace=True)
df
```

Our output looks good. Next, we'll convert the Trans, Drive, Fuel, Veh Class, and SmartWay columns to Integer representations.

We can do this by utilizing dictionaries and arrays. First we put the unique column values into an array. We assign each unique column value the array index Integer in the dictionary. Then we simply map the DataFrame to the dictionary.

For example, in tr_dict (Trans dictionary), for the key 'AMS-8', the value is 2.

```python
tr_dict = {}
dr_dict = {}
fu_dict = {}
vc_dict = {}
sw_dict = {}

tr_arr = df['Trans'].unique()
dr_arr = df['Drive'].unique()
fu_arr = df['Fuel'].unique()
vc_arr = df['Veh Class'].unique()
sw_arr = df['SmartWay'].unique()

for i in range(tr_arr.size):
    tr_dict[tr_arr[i]] = i

for i in range(dr_arr.size):
    dr_dict[dr_arr[i]] = i
    
for i in range(fu_arr.size):
    fu_dict[fu_arr[i]] = i
    
for i in range(vc_arr.size):
    vc_dict[vc_arr[i]] = i
    
for i in range(sw_arr.size):
    sw_dict[sw_arr[i]] = i
    
df['Trans'] = df['Trans'].map(tr_dict)
df['Drive'] = df['Drive'].map(dr_dict)
df['Fuel'] = df['Fuel'].map(fu_dict)
df['Veh Class'] = df['Veh Class'].map(vc_dict)
df['SmartWay'] = df['SmartWay'].map(sw_dict)

df
```

<b>Step 2: Replace all N/A with medians for all numeric columns except the ones in the previous line: for those choose mode.</b>

Lets import NumPy and start finding the N/A values.

```python
import numpy as np
df = df.replace('?', np.NaN)

df.isnull().sum()
```

It looks like the only 2 columns that have N/A (or NaN rather) values are Model and Displ.

First we look for all of the numeric columns except the ones in the previous line - I can only assume that this step is referring to the 5 columns in Step 1: Trans, Drive, Fuel, Veh Class, and SmartWay.

Since Model and Displ weren't converted in Step 1, <u>and</u> Model and Displ are numeric columns, <b>we will proceed by replacing the N/A in the Model and Displ columns with their medians.</b>



```python
dl_median = df['Displ'].median()
cl_median = df['Cyl'].median()

# replace the NaN in each column with the median label
df.loc[df['Displ'].isnull(), 'Displ'] = dl_median
df.loc[df['Cyl'].isnull(), 'Cyl'] = cl_median
```

Verify that the NaN values are replaced by their respective column median value.
