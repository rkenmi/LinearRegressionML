<b>Step 1: Convert Trans, Drive, Fuel, Veh Class, SmartWay column values to integers.</b><br/>

First we will import pandas and read the csv file.

```python
import pandas as pd
df = pd.read_csv("milage.csv", skipinitialspace=True)
df
```

Our output looks good. 

Next, we'll convert the Trans, Drive, Fuel, Veh Class, and SmartWay columns to Integer representations.

We can do this by utilizing dictionaries and arrays. First we put the unique column values into an array. We assign each unique column value the array index Integer in the dictionary. Then we simply map the DataFrame to the dictionary.

For example, in tr_dict (Trans dictionary), for the key 'AMS-8', the value is 2.

```python
tr_dict = {}
dr_dict = {}
fu_dict = {}
vc_dict = {}
sw_dict = {}

tr_arr = df['Trans'].unique()
dr_arr = df['Drive'].unique()
fu_arr = df['Fuel'].unique()
vc_arr = df['Veh Class'].unique()
sw_arr = df['SmartWay'].unique()

for i in range(tr_arr.size):
    tr_dict[tr_arr[i]] = i

for i in range(dr_arr.size):
    dr_dict[dr_arr[i]] = i
    
for i in range(fu_arr.size):
    fu_dict[fu_arr[i]] = i
    
for i in range(vc_arr.size):
    vc_dict[vc_arr[i]] = i
    
for i in range(sw_arr.size):
    sw_dict[sw_arr[i]] = i
    
df['Trans'] = df['Trans'].map(tr_dict)
df['Drive'] = df['Drive'].map(dr_dict)
df['Fuel'] = df['Fuel'].map(fu_dict)
df['Veh Class'] = df['Veh Class'].map(vc_dict)
df['SmartWay'] = df['SmartWay'].map(sw_dict)

df
df.dtypes
```

<b>Step 3: Implement scale_features as per lecture, it should take one training example and return a vector of scaled and mean-normalized values.</b>

First off, since y represents the 'Cmb MPG' column, we need to convert this column to numbers. By default, this column is an Object type because there are certain y values with two numbers and a slash character in between.

Thus we need to eliminate the slash and add the two numbers together and divide by 2.

Then we'll convert the whole column to integers.

```python
if 'Cmb MPG' in df.columns:
    slashIndices = df[ df['Cmb MPG'].str.contains('/') ]['Cmb MPG'].index

    for i in slashIndices:
        if(df['Cmb MPG'][i]):
            df.loc[i,'Cmb MPG'] = (int(df['Cmb MPG'][i].split('/')[0]) + int(df['Cmb MPG'][i].split('/')[1])) / 2

    df['Cmb MPG'] = df['Cmb MPG'].astype('int64')
```

For feature scaling, the maximum and minimum values must be found for each column. To produce mean-normalized values, we will of course need the mean values for each feature as well.

```python
if 'Cmb MPG' in df.columns:
    yDf = df['Cmb MPG'] # save y dataframe for later
    df = df.drop('Cmb MPG', 1) # drop y, since we will be treating this dataframe as X
    data = df.select_dtypes(['float64', 'int64']) # this will filter df out, leaving only the FEATURES in data

aMax = data.values.max(axis=0) # get the max of each column
aMin = data.values.min(axis=0) # get the min of each column
aMean = data.values.mean(axis=0) # get the mean of each column

def scale_features(training_example):
    return (training_example - aMean) / (aMax - aMin)
```

Then we will update the old columns in the dataframe accordingly using this newly created matrix with a simple loop.

```python
scaled_data = scale_features(data.values)

i = 0
for value in data.columns.values:
    df[value] = scaled_data[:, i]
    i += 1
    
df
```

<b>Step 6: Implement cost_function as per lecture/Ng videos: it takes the above numpy matrix and returns a cost value.</b>

We will use the y_train value from the global scope since this function only takes the X matrix.

Here, I assume that the theta values will also be supplied from the global scope. (I set the thetas to 0s as sample values.)

```python
def cost_function(X_matr, y_matr):
    
    hypo = theta.dot(X_matr.T)
    
    diff = hypo - y_matr
    sqErr = diff.T.dot(diff)
    m = y_matr.size
    
    return (1.0 /(2 * m)) * sqErr.sum()

theta = np.zeros((1, X_matr[0, :].size)) # (theta_0, theta_1, ... theta_n)
cost_function(X_matr, y_matr)
```

<b>Step 7: Implement gradient as per lecture/Ng videos.</b>

gradient takes 3 parameters. This function will be used in the gradient_descent function.

*hypo* is the mathematical function for h_theta (x^i). 

*y* is the y vector. 

*x_hat_i* or *x^i* is all of the rows of the current n_th feature in the X matrix.

```python
def gradient(hypo, y, x_hat_i):
    m = y.size
    return (1.0 / m) * ((hypo - y) * x_hat_i).sum()
```

<b>Step 8: Implement gradient_descent as per lecture/Ng videos. It should return a vector of trained parameters. It should take in a numpy array, num_iter (number of iterations),  and an alpha (learning rate).</b>

Note:
- The function takes 4 parameters, but the theta values are required (from global scope). I assumed that the theta values are provided outside the scope of gradient_descent based on the instructions, hence the *global theta*



```python
X_arr = np.array(X_matr) # feature-scaled X matrix array
y_arr = np.array(y_matr) # Cmb MPG

def gradient_descent(X, y, num_iter, alpha):
    global theta # so we don't save to a "local" theta, but to the "global" theta
    plot_arr = []
    
    n = X.shape[1] # n is the amount of columns or features (x_0, x_1, x_2, ... x_n)
    
    hypo = theta.dot(X.T)
    
    for i in range(num_iter):
        for j in range(n):
            theta[0, j] = theta[0, j] - (alpha * gradient(hypo, y, X[:, j])) # save to "global" theta
        plot_arr.append(cost_function(X_matr, y_matr))

    return plot_arr, theta
```
